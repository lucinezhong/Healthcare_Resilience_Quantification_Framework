This aims to evaluate the flexibility of the beta family equation in capturing versatile amplitude, disruption, recovery rate, and duration.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def beta_function(t,a,b,c,v):
    return a*(math.pow((b+c),b+c)/(math.pow(b,b)*math.pow(c,c)))*np.power(t/v,b)*np.power(1-t/v,c)
```
